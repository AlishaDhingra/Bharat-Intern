```python
# Convert the 'label' column to a binary classification problem (spam/ham)
df['label'] = np.where(df['v1'] == 'pam', 1, 0)
```

```python
#drop unnecessary columns from database
df.drop(['Unnamed: 2',	'Unnamed: 3' ,	'Unnamed: 4'], axis=1, inplace=True)
```

```python
df
```

```python
# Calculate the proportion of spam and ham messages
spam_proportion = df['label'].mean()
ham_proportion = 1 - spam_proportion
print(f'Spam Proportion: {spam_proportion:.2f}')
print(f'Ham Proportion: {ham_proportion:.2f}')
```

```python
# Calculate the average message length for spam and ham messages
avg_spam_length = df[df['label'] == 1]['v2'].apply(len).mean()
avg_ham_length = df[df['label'] == 0]['v2'].apply(len).mean()
print(f'Average Spam Message Length: {avg_spam_length:.2f}')
print(f'Average Ham Message Length: {avg_ham_length:.2f}')
```

```python
# Plot the distribution of spam and ham messages
sns.countplot(x='v1', data=df)
plt.xlabel('Label')
plt.ylabel('Frequency')
plt.title('Spam vs Ham Messages')
plt.show()
```

```python
# Plot the length of spam and ham messages
plt.figure(figsize=(10, 6))
sns.boxplot(x='v1', y='message_length', data=df.assign(message_length=df['v2'].apply(len)))
plt.xlabel('Label')
plt.ylabel('Message Length')
plt.title('Message Length Distribution')
plt.show()
```
